```python
data['TotalCharges'] = data['TotalCharges'].astype(str).str.strip()

data['TotalCharges'].replace('', pd.NA, inplace=True)

data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')

data['TotalCharges'].fillna(data['TotalCharges'].median(), inplace=True)

```

Now i will Scale the following columns - tenure, monthly charges, total charges

```python
import sklearn 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data['tenure'] = scaler.fit_transform(data[['tenure']])
data['MonthlyCharges'] = scaler.fit_transform(data[['MonthlyCharges']])
data['TotalCharges'] = scaler.fit_transform(data[['TotalCharges']])
```

```python
data.head()
```

Dropping columns which are not necessory

```python
drop_columns = ['customerID', 'PaperlessBilling', 'PaymentMethod']
data.drop(drop_columns, axis=1, inplace=True)
```

```python
data
```

Now i will convert Categorical columns into numerical.

```python
Categorical_columns = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'Churn' ]
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse_output=False)
data['gender'] = encoder.fit_transform(data[['gender']])
```

```python
encoded_columns = encoder.fit_transform(data[Categorical_columns])
encoded_df = pd.DataFrame(encoded_columns, columns=encoder.get_feature_names_out(Categorical_columns))
encoded_df
```

Replacing Categorical columns with encoded data

```python
data = pd.concat([data.drop(columns = Categorical_columns), encoded_df], axis=1)
```

```python
from sklearn.model_selection import train_test_split
train, temp = train_test_split(data, test_size=0.4, random_state=42)
val, test = train_test_split(data, random_state=42, train_size=0.5)
```
