```python
import pandas as pd

xgboost = pd.read_csv('xgboost_model_comparison.csv')
xgboost
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(9,5))
x = np.arange(len(xgboost['Model Number']))
width = 0.35
plt.bar(x - width/2, xgboost['Accuracy score Validation']*100, width, label='Validation', alpha=0.8)
plt.bar(x + width/2, xgboost['Accuracy score Test']*100, width, label='Test', alpha=0.8)
plt.xticks(x, xgboost['Model Number'], rotation=30, ha='right')
plt.ylabel('Accuracy (%)')
plt.ylim(0, 100)
plt.title('Model Accuracy Comparison (Validation vs Test)')
plt.legend()
plt.tight_layout()
plt.show()
```

Tune model-3 is selected as final XGBoost model, Reason- it offers best balanced validation and test accuracy.

```python
plt.figure(figsize=(9,5))
x = np.arange(len(xgboost['Model Number']))
width = 0.35
plt.bar(x - width/2, xgboost['Precision Score Validation']*100, width, label='Validation', alpha=0.8)
plt.bar(x + width/2, xgboost['Precision Score Test']*100, width, label='Test', alpha=0.8)
plt.xticks(x, xgboost['Model Number'], rotation=30, ha='right')
plt.ylabel('Precision Score (%)')
plt.ylim(0, 100)
plt.title('Model Precision Score Comparison (Validation vs Test)')
plt.legend()
plt.tight_layout()
plt.show()
```

Tune model-3 is selected as final XGBoost model,Reason - it provides a balanced precision between validation and test datasets.

```python
plt.figure(figsize=(9,5))
x = np.arange(len(xgboost['Model Number']))
width = 0.35
plt.bar(x - width/2, xgboost['Recall Validation']*100, width, label='Validation', alpha=0.8)
plt.bar(x + width/2, xgboost['Recall Test']*100, width, label='Test', alpha=0.8)
plt.xticks(x, xgboost['Model Number'], rotation=30, ha='right')
plt.ylabel('Recall Score (%)')
plt.ylim(0, 100)
plt.title('Model Recall Score Comparison (Validation vs Test)')
plt.legend()
plt.tight_layout()
plt.show()
```

Tune model-3 is selected as final XGBoost model, Reason- it maintains the highest and most balanced recall across validation and test datasets.

```python
plt.figure(figsize=(9,5))
x = np.arange(len(xgboost['Model Number']))
width = 0.35
plt.bar(x - width/2, xgboost['F1 Score Validation']*100, width, label='Validation', alpha=0.8)
plt.bar(x + width/2, xgboost['F1 Score Test']*100, width, label='Test', alpha=0.8)
plt.xticks(x, xgboost['Model Number'], rotation=30, ha='right')
plt.ylabel('F1 Score (%)')
plt.ylim(0, 100)
plt.title('Model F1 Score Comparison (Validation vs Test)')
plt.legend()
plt.tight_layout()
plt.show()
```

Tune model-3 is selected as final XGBoost model, Reason- t provides a strong and balanced F1 score across validation and test datasets.

```python
model3 = xgboost[xgboost['Model Number'] == 'Model 3'].iloc[0]


metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
validation_scores = [
    model3['Accuracy score Validation'],
    model3['Precision Score Validation'],
    model3['Recall Validation'],
    model3['F1 Score Validation']
]
test_scores = [
    model3['Accuracy score Test'],
    model3['Precision Score Test'],
    model3['Recall Test'],
    model3['F1 Score Test']
]


x = np.arange(len(metrics))
width = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - width/2, np.array(validation_scores)*100, width, label='Validation', alpha=0.8)
plt.bar(x + width/2, np.array(test_scores)*100, width, label='Test', alpha=0.8)

plt.xticks(x, metrics)
plt.ylabel('Score (%)')
plt.ylim(0, 100)
plt.title('Model 3  Validation vs Test Metrics Comparison')
plt.legend()
plt.tight_layout()
plt.show()
```

Model 3 shows a strong balance across all key performance metrics for both validation and test data.
It maintains high recall, ensuring most churn cases are correctly identified, while keeping good precision.
The minimal gap between validation and test scores proves its consistency and generalization.
Overall, Model 3 is a stable and reliable model for predicting customer churn in real-world scenarios.
